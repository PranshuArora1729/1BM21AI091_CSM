```python
import numpy as np
import matplotlib.pyplot as plt

def parabola(x, a, b, c):
    return -a * x**2 + b * x + c

def parabola_derivative(x, a, b):
    return 2 * a * x + b

def tangent_line(x, x0, a, b, c):
    y0 = parabola(x0, a, b, c)
    slope = parabola_derivative(x0, a, b)
    return slope * (x - x0) + y0

def draw_parabola(a, b, c, x_range=(-10, 10), num_points=100):
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = parabola(x_values, a, b, c)

    plt.plot(x_values, y_values, label=f'Parabola: {a}x^2 + {b}x + {c}')

def draw_tangents(a, b, c, x_range=(-10, 10), num_tangents=5):
    x_values = np.linspace(x_range[0], x_range[1], 1000)
    step = (x_range[1] - x_range[0]) / (num_tangents + 1)

    for i in range(1, num_tangents + 1):
        x_tangent = x_range[0] + i * step
        y_tangent = tangent_line(x_values, x_tangent, a, b, c)
        plt.plot(x_values, y_tangent, '--', label=f'Tangent at x={x_tangent:.2f}')


a = 0.15646
b = 0.56789
c = 2


draw_parabola(a, b, c)


draw_tangents(a, b, c)


plt.xlim(-20, 20)
plt.ylim(-10, 10)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Parabola with Tangents')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
